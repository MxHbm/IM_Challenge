```python
def is_route_feasible(route, distance_matrix, tasks):
    current_time = 0  # Start time from the depot
    for i in range(len(route) - 1):
        current_node = route[i]
        next_node = route[i + 1]
        print(f"\nTraveling from {current_node} to {next_node}")
        print(f"Current time before traveling: {current_time}")
        
        # Travel time to the next node
        travel_time = distance_matrix[current_node][next_node]
        current_time += travel_time
        print(f"Current time after traveling: {current_time}")
        
        # Check if the next node is not the depot (0) or the end point (15)
        if next_node != 0 and next_node != 15:
            task_start_time = tasks[next_node]['Start Time']
            service_time = tasks[next_node]['Service Time']
            print(f"Next node start time: {task_start_time}, service time: {service_time}")
            
            # Check if current time is within the start time of the next node
            if current_time < task_start_time:
                print(f"Waiting until start time of this node: {task_start_time-current_time}")
                current_time = task_start_time  # Wait until the start time
            
            current_time += service_time  # Add the service time
            print(f"Current time after service: {current_time}")
            
            if current_time > 21600:
                print("Not feasible: exceeded maximum allowed time")
                return False

    return current_time <= 21600  # Check if the route finishes within the maximum allowed time

def check_solution_feasibility(solution, distance_matrix, tasks):
    for day, cohorts in solution.items():
        print("________________________________________________________________________________")
        for cohort, route in cohorts.items():
            print("________________________________________________________________________________")
            print(f"Checking feasibility for day {day}, cohort {cohort}, route {route}")
            if not is_route_feasible(route, distance_matrix, tasks):
                return False
    return True


# Check feasibility
is_feasible = check_solution_feasibility(solution, distance, mainTasks)
print("\nSolution is feasible:", is_feasible)
```
