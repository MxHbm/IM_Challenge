```python
# Create DataFrame

import pandas as pd

import matplotlib.pyplot as plt

csv = pd.read_csv('profit_over_time_7_8_1.csv')

df = pd.DataFrame(csv)

# Plot the graph
plt.figure(figsize=(14, 4))
plt.plot(df['RunTime'], df['Profit'], marker='o', color='b', label='Profit over Time')
plt.title('Profit over RunTime')
plt.xlabel('RunTime (seconds)')
plt.ylabel('Profit')
plt.grid(True)
plt.legend()

# choose layout
plt.tight_layout()


# Show the plot
plt.show()
```
