```python
# Create DataFrame

import pandas as pd

import matplotlib.pyplot as plt

csv = pd.read_csv('7_2_1_profit_over_time.csv')

df = pd.DataFrame(csv)

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(df['RunTime'], df['Profit'],  color='b', label='Profit over Time')
plt.title('Profit over RunTime')
plt.xlabel('RunTime (seconds)')
plt.ylabel('Profit')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Berechne die Differenz der Profit-Spalte
df['Profit_Change'] = df['Profit'].diff()

# Filtere die Zeitpunkte heraus, an denen sich der Profit geändert hat
profit_changes = df[df['Profit_Change'] != 0][['RunTime', 'Profit_Change']]

# Zeige die Zeitpunkte und die Stärke der Änderungen
profit_changes['Previous_Profit'] = profit_changes['RunTime'].apply(lambda x: df.loc[df['RunTime'] == x, 'Profit'].values[0] - df.loc[df['RunTime'] == x, 'Profit_Change'].values[0])
profit_changes['New_Profit'] = profit_changes['RunTime'].apply(lambda x: df.loc[df['RunTime'] == x, 'Profit'].values[0])
print(profit_changes)
```
